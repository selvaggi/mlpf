```python
import wandb
import pandas as pd
from datetime import datetime

def convert_wandb_date(s):
    return datetime.strptime(s, '%Y-%m-%dT%H:%M:%SZ')


# Login to wandb
wandb.login()

# Specify your project details
ENTITY = "fcc_ml"  # Replace with your WandB entity
PROJECT = "mlpf_debug"  # Replace with your WandB project name

# Fetch runs from the project
api = wandb.Api()
#runs = api.runs(f"{ENTITY}/{PROJECT}")
# only select runs from July to Oct 2024
runs = api.runs(f"{ENTITY}/{PROJECT}", filters={"created_at": {"$gte": "2024-07-01", "$lt": "2024-11-01"}})

i = 0
# Extract the desired information
data = []
for run in runs:
    i+=1
    if i % 10==0:
        print(i)
    data.append({
        "wandb_name": run.name,
        "human_readable_date": run.created_at,
        "link": run.url,
        "command": " ".join(run.metadata['args']),  # Default to 'N/A' if no program is specified
    })

# Convert to pandas DataFrame
df = pd.DataFrame(data)
# Save the DataFrame to a CSV if needed
#df.to_csv("/eos/user/g/gkrzmanc/results/2024/wandb_runs_mlpf_debug_2024.csv", index=False)
# save also to pickle
df.to_pickle("/eos/user/g/gkrzmanc/results/2024/wandb_runs_mlpf_debug_2024.pkl")
# Display the DataFrame
print(df)
```
